```python
#수정
Top_hightop_train = pd.DataFrame(columns=["image_file_name","color","Style","Part","Category"])

j = []
for i in range(1,41):
    file_name = "Top_{}.jpg".format(i) #수정
    j.append(file_name)

Top_hightop_train["image_file_name"]=j #수정
Top_hightop_train["Part"]="Top" #수정
Top_hightop_train["Category"]="Top_hightop" #수정


########################################

#수정
Top_hightop_test = pd.DataFrame(columns=["image_file_name","color","Style","Part","Category"])

j = []
for i in range(41,51):
    file_name = "Top_{}.jpg".format(i) #수정
    j.append(file_name)

Top_hightop_test["image_file_name"]=j #수정
Top_hightop_test["Part"]="Top" #수정
Top_hightop_test["Category"]="Top_hightop" #수정

a = Top_hightop_train.to_csv("Top_hightop_train.csv",mode = "w", sep=',') #수정
a1 = Top_hightop_test.to_csv("Top_hightop_test.csv",mode = "w", sep=',') #수정
```
